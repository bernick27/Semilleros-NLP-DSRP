# Extracción de texto de los capítulos del libro "Frankenstein"

```python
r = requests.get('https://www.gutenberg.org/files/84/84-h/84-h.htm')
soup = BeautifulSoup(r.text, 'html.parser')
```

```python
soup.prettify()
```

```python
contenido = []
for seccion in soup.find_all('div'):
    if seccion.h2 is not None:
      if re.search('Chapter\s\d\d?', seccion.h2.text) is not None:
        contenido.append(seccion.h2.text)
        for parrafo in seccion.find_all('p'):
          contenido.append(parrafo.text)

for c in contenido:
  print(c)
```

# Limpieza o filtrado de los textos

```python
def limpieza_texto(texto):
  texto = re.sub(r'Ã¦', 'ae', texto)
  texto = re.sub(r'[\[\]]', '', texto)
  texto = re.sub(r'[\.|:|,|;|!][”|’]?\s', ' ', texto)
  texto = re.sub(r'(?P<ref1>(\w))\s\n(?P<ref2>(\w))', '\g<ref1> \g<ref2>', texto)
  texto = re.sub(r'\s\n', '', texto)
  texto = re.sub(r'[,;!“”?’‘\(\)]', '', texto)
  texto = re.sub(r'[—-]', ' ', texto)
  texto = re.sub(r'’[^\s]+', '', texto)
  texto = re.sub(r'\s\s{1,}', ' ', texto)
  texto = re.sub(r'[\.|:]', '', texto)
  
  return texto.lower()
```

```python
txt = limpieza_texto(contenido[len(contenido)-22])
```

```python
txt
```

```python
contenido_filtrado = []
for parrafo in contenido:
  contenido_filtrado.append(limpieza_texto(parrafo))

for c in contenido_filtrado:
  print(c)
```

# Tokenización antes de stemming

```python
import nltk
nltk.download('punkt')
from nltk.tokenize import word_tokenize, sent_tokenize
from nltk.corpus import stopwords
nltk.download('stopwords')

def frecuencia_tokens(tokens_unico, list_textos):
  dict_freq_tokens = dict()
  for token in tokens_unico:
    dict_freq_tokens[token] = 0
  for txt in list_textos:
    try:
      for term in txt.split(' '):
        try:
          dict_freq_tokens[term] += 1
        except:
          pass
    except:
      for term in txt:
        try:
          dict_freq_tokens[term] += 1
        except:
          pass
  return dict_freq_tokens

contenido_tokenizado = []
for i in range(len(contenido_filtrado)):
  contenido_tokenizado.append(word_tokenize(contenido_filtrado[i]))

# Haciendo flatten la lista contenido_tokenizado
contenido_tokenizado_flatten = [item for sublist in contenido_tokenizado for item in sublist]

# Hallando palabras unicas en lista contenido_tokenizado_flatten
contenido_tokenizado_unico = list(set(contenido_tokenizado_flatten))

print(contenido_tokenizado_unico)
```

```python
print('Total palabras tokenizadas antes de stemming (tamaño de vocabulario): ', len(contenido_tokenizado_unico))
```

```python
stop_words = stopwords.words('english')
contenido_tokenizado_unico_sin_stopwords = [word for word in contenido_tokenizado_unico if word not in stop_words]
diccionario_tokens_sin_stemming = frecuencia_tokens(contenido_tokenizado_unico_sin_stopwords, contenido_filtrado)
sorted_dict_sin_stemming = {k: v for k, v in sorted(diccionario_tokens_sin_stemming.items(), key=lambda item: item[1], reverse=True)}
```

```python
print('Términos relevantes más frecuentes en la lista de Tokens:\n', sorted_dict_sin_stemming)
```

# Aplicación de Stemming: Algoritmo de Porter

```python
from nltk.stem.porter import PorterStemmer
porter = PorterStemmer()

# tokenized_words = contenido_filtrado[1].split(' ')
# stemm_ejemplo = [porter.stem(word) for word in tokenized_words]
# print(stemm_ejemplo)

contenido_porter = []
contenido_stemm = []
for i in range(len(contenido_filtrado)):
  contenido_porter.append(contenido_filtrado[i].split(' '))
  contenido_stemm.append( [porter.stem(word) for word in contenido_porter[i]] )

# Haciendo flatten la lista contenido_stemm
contenido_stemm_flatten = [item for sublist in contenido_stemm for item in sublist]

# Hallando palabras unicas en lista contenido_stemm_flatten
contenido_stemm_unico = list(set(contenido_stemm_flatten))

print(contenido_stemm_unico)
```

```python
print('Total palabras tokenizadas después de stemming (tamaño de vocabulario): ', len(contenido_stemm_unico))
```

```python
stop_words = stopwords.words('english')
contenido_stemm_unico_sin_stopwords = [word for word in contenido_stemm_unico if word not in stop_words]
diccionario_tokens_con_stemming = frecuencia_tokens(contenido_stemm_unico_sin_stopwords, contenido_stemm)
sorted_dict_con_stemming = {k: v for k, v in sorted(diccionario_tokens_con_stemming.items(), key=lambda item: item[1], reverse=True)}
```

```python
print('Términos relevantes más frecuentes en la lista de Tokens:\n', sorted_dict_con_stemming)
```
